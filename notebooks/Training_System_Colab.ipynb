## ğŸ“Š æ­¥é©Ÿ 7: ç¯„ä¾‹è¨“ç·´æµç¨‹

```python
# ç¯„ä¾‹: å®Œæ•´è¨“ç·´æµç¨‹ç¤ºç¯„

print("=" * 60)
print("ç¤ºç¯„: å®Œæ•´è¨“ç·´æµç¨‹")
print("=" * 60)

# 1. ç”Ÿæˆæƒ…å¢ƒ
scenario = generator.generate_scenario(difficulty="MEDIUM")
print(f"\n1. æƒ…å¢ƒç”Ÿæˆ")
print(f"   é¡å‹: {scenario['type']}")
print(f"   é›£åº¦: {scenario['difficulty']}")
print(f"   åç¨±: {scenario['template']['name']}")

# 2. æ³¨å…¥æ•…éšœ
twin.inject_fault(scenario['type'])
state = twin.export_current_state()
print(f"\n2. æ•…éšœæ³¨å…¥å®Œæˆ")

# 3. A2A å°ˆå®¶è¨ºæ–·
diagnosis = coordinator.start_diagnosis_session(state, student_level="beginner")
print(f"\n3. å°ˆå®¶è¨ºæ–·å®Œæˆ")
print(f"   è¨ºæ–·çµæœ: {diagnosis['session_summary']['fault_type']}")
print(f"   ä¿¡å¿ƒåº¦: {diagnosis['session_summary']['confidence']:.2%}")

# 4. è©•åˆ†ï¼ˆæ¨¡æ“¬ï¼‰
from evaluation.scoring_system import ScoringSystem
scorer = ScoringSystem()

result = scorer.evaluate_session(
    student_id="DEMO_STUDENT",
    scenario=scenario,
    student_actions=[
        {"action": "æª¢æŸ¥è¨­å‚™", "is_correct": True},
        {"action": "åŸ·è¡Œä¿®å¾©", "is_correct": True}
    ],
    diagnosis_result=diagnosis,
    completion_time=20.0
)

print(f"\n4. è©•åˆ†çµæœ")
print(f"   ç¸½åˆ†: {result['scores']['total']:.1f}")
print(f"   ç­‰ç´š: {result['grade']}")
print(f"\nâœ… ç¤ºç¯„å®Œæˆï¼")
```
